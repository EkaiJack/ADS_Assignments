# Sample Socio-Demographic Effect on Attrition

```python

age_effect = df.groupby('Attrition').mean()['Age']
age_effect
#The average age of employees who are lost by attrition is 33.61
```

```python
dist_effect = df.groupby('Attrition').mean()['DistanceFromHome']
dist_effect
# The average distance from home for those who exited the company through attrition was higher at (10.63)
```

```python
MonthlyIncome_effect = df.groupby('Attrition').mean()['MonthlyIncome']
MonthlyIncome_effect
  #Employees who exited the company through atrrition had a lower average monthly Income.
```

```python
GenderEffect = pd.crosstab(df['Gender'],df['Attrition'], margins = False, normalize='columns')
GenderEffect
#Larger percentage of those who left the company are Male Employees.
```

# A breakdown of distance from home by job role and attrition.

# Comparison of average monthly income by education and attrition.

```python
df.groupby(["Education","Attrition"]).mean() ["MonthlyIncome"]
```
